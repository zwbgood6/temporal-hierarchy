```python
import gym
import gym_push
import numpy as np
import os
import os.path as osp
from matplotlib import pyplot as plt
import cv2
```

```python
env = gym.make('CustomPush-v0')
env.seed(42)
env.reset()
ac_dim = env.action_space.shape[0]
```

```python
rendered_image = env.render(mode='rgb_array')
plt.imshow(rendered_image)
```

```python
n_step = 100
```

```python
for i in range(n_step):
    ac = np.array([0, 0, 0, 0, 0, 0]) if i % 100 == 0 else np.zeros((ac_dim,))
    obs, rew, done, _ = env.step(ac)
rendered_image = env.render(mode='rgb_array')
plt.imshow(rendered_image)
```
